```python
df.drop(columns = ['id','Unnamed: 32'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_test,X_test.shape)
```

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
print(y_test,y_test.shape,type(y_test))
```

```python
import torch
X_train = torch.from_numpy(X_train)
X_test = torch.from_numpy(X_test)
y_train = torch.from_numpy(y_train)
y_test = torch.from_numpy(y_test)
print(y_test,y_test.shape,type(y_test),y_test.dtype)
```

#3.Training Process

```python
class onelayernn:

  def __init__(self,X):
    self.weights = torch.rand(X.shape[1],1, dtype = torch.float64, requires_grad=True)
    self.bias = torch.rand(1,dtype=torch.float64, requires_grad=True)

  def forward_pass(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss(self,y_pred):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon) #clamping to avoid log(0) after sigmoid operation
    loss = -((y_train * torch.log(y_pred)) + ((1-y_train)*torch.log(1-y_pred))).mean()
    return loss
```

```python
epochs = 250
learning_rate = 0.1
```

```python
#A. Creating Model
model = onelayernn(X_train)

for epoch in range(epochs):
  #B. Forward Pass
  y_pred = model.forward_pass(X_train)
  #C.Loss Calculation
  loss = model.loss(y_pred)
  print(f"Loss for Epoch {epoch+1}: ",loss.item())
  loss.backward()
  #D.Parameter Updates using Backpropogation
  with torch.no_grad():
    model.weights -= model.weights.grad * learning_rate
    model.bias -= model.bias.grad * learning_rate

  model.weights.grad.zero_()
  model.bias.grad.zero_()
```

#4.Evaluation

```python
with torch.no_grad():
  y_test_pred = model.forward_pass(X_test)
  y_test_pred = (y_test_pred>0.5).float()
  accuracy = (y_test_pred == y_test).float().mean()
  print("Accuracy: ", accuracy.item())
```
